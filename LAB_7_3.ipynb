```python
import numpy as np
w, b = np.array([0,0]),0.0
learning_rate = 0.01

def activation(s):
  if s > 0: return 1
  elif s < 0: return -1
  return 0

def out(x) : 
  return activation (w.dot(x) + b)

def train(x0, x1, target):
  global w, b
  x = np.array([x0, x1])
  y = out(x)

  ##예측이 맞으면 아무것도 안함
  if target == y: return False
  ##예측이 틀리면 학습      ##가중치 변경 X
  print('가중치 수정전 target :{} y:{} b:{} w:{}'.format(target, y ,b, w))

  w = w +learning_rate * x * target
  b = b +learning_rate * 1 * target
  print('가중치 수정후 target :{} y:{} b:{} w:{}'.format(target, y ,b, w))
  return True

def predict(inputs):
  outputs = []
  for x in inputs:
    outputs.append (out(x))
  return outputs
```

```python
adjusted_OR = 0
for i in range(100):
  adjusted_OR += train(-1,-1, -1)
  adjusted_OR += train(-1, 1,  1)
  adjusted_OR += train( 1,-1,  1)
  adjusted_OR += train( 1, 1,  1)
  print("iteration -----------", i)
  if not adjusted_OR: break
  adjusted_OR = 0 
```

```python
x = [[-1,-1],[-1,1],[1,-1],[1,1]]
yhat = predict(x)
print('x0 x1  y')
for i in range(len(x)):
  print('{0:2d} {1:2d} {2:2d}'.format(x[i][0], x[i][1], yhat[i]))
```
